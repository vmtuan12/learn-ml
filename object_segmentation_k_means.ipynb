```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
```

```python
img = mpimg.imread('girl3.jpg')
plt.axis(False)
imgplot = plt.imshow(img)
```

```python
# Create matrix X that each row represents a pixel with 3 colors
# img.shape[0] corresponds to the height and img.shape[1] corresponds to the width -> number of pixels of the img
# img.shape[2] represents the number of channels, which could be 3 for RGB images
X = img.reshape((img.shape[0]*img.shape[1], img.shape[2]))
print(X)
```

```python
for K in [2, 3, 5, 10, 20, 100]:
    k_means = KMeans(n_clusters=K).fit(X)
    pred_labels = k_means.predict(X)

    img4 = np.zeros_like(X)
    # replace each pixel by its center
    for k in range(K):
        img4[pred_labels == k] = k_means.cluster_centers_[k]

    img5 = img4.reshape((img.shape[0], img.shape[1], img.shape[2]))
    plt.imshow(img5)
    plt.axis(False)
    plt.show()
```
